# **Image to Pencil Sketch with Python:**
**Importing Libraries**

```python
import cv2
import matplotlib.pyplot as plt
```

# **Loading the original img**

```python
img_sketch = cv2.imread("/content/image.jpg")
```

# **Showing original img in RGB Format**

```python
img_sketch = cv2.cvtColor(img_sketch,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(img_sketch)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# **Converting Image to GrayScale**

```python
img_sketch_gray = cv2.cvtColor(img_sketch, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(img_sketch_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

# **Inverting the grayscale image**

```python
img_sketch_invert = cv2.bitwise_not(img_sketch_gray)
plt.figure(figsize=(7,7))
plt.imshow(img_sketch_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

# **Smooting img**
remove edges and make it smooth for sketch

```python
img_smoothing = cv2.GaussianBlur(img_sketch_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(7,7))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

# **Final Image to Pencil Sketch**

```python
final_pencil_sketch = cv2.divide(img_sketch_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(7,7))
plt.imshow(final_pencil_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

# **Final Output**
Displaying all the images in one frame using the subplots.

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img_sketch)
plt.axis("off")
plt.title("Original Image with RGB format")
plt.subplot(1,5,2)
plt.imshow(img_sketch_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_sketch_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final_pencil_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```
